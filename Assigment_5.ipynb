```python
import requests
import pandas as pd
import datetime 

date = datetime.date.today() + datetime.timedelta(days=1)

url = f"https://api.open-meteo.com/v1/forecast?latitude=50.4547&longitude=30.5238&hourly=temperature_2m,wind_speed_10m&timezone=auto&start_date={date}&end_date={date +datetime.timedelta(days=6)}"
data = requests.get(url).json()
time = pd.Series(data["hourly"]["time"])
temperature = pd.Series(data["hourly"]["temperature_2m"])
wind_speed = pd.Series(data["hourly"]["wind_speed_10m"])
dict = {
   "temperature": temperature,
    "wind_speed": wind_speed,
    "time":time,
} 
is_wind_speed_more_avg = pd.Series([dict["wind_speed"] > (sum(dict["wind_speed"]) / len(dict["wind_speed"]))])[0]
dict["is_wind_speed_more_avg"] = is_wind_speed_more_avg
df = pd.DataFrame(dict)
display(df)

print("A number of hours when wind speed more than average wind speed :", sum(df["is_wind_speed_more_avg"]), end="\n\n")

print("min for 3 days :", min(df["temperature"].loc[:71]))
print("max for 3 days :", max(df["temperature"].loc[:71]))
print("avg for 3 days :", sum(df["temperature"].loc[:71]) / 72)

```
